## **target(continues) vs feature(countinues) ----> scatter**


````python
def scatter_plot(data_frame, feature, target):
        plt.figure()
        plt.scatter(x=data_frame[feature], y=data_frame[target])
        plt.xlabel(feature)
        plt.ylabel(target)
        plt.show()
        print(f"{feature} mean: ", data_frame[feature].mean())
        print(f"{feature} std mean: ", data_frame[feature].std())
        print(f"{feature} median: ", data_frame[feature].median())  
````

## **scaling features to a range**

Bring all features in the same standing, so that one significant number doesn’t impact the model just because of their large magnitude



*   Standard Scaling

    z = (x - u) / s
*   MinMaxAcaler

    X_std = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))

    X_scaled = X_std * (max - min) + min



````python
X = boston_df.iloc[:, :-1]
y = boston_df.iloc[:,-1]
X.shape
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.4 , random_state=42)
X_train.shape
````

````python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
standard_scaler = preprocessing.StandardScaler()
X_train_scaled = standard_scaler.fit_transform(X_train, y_train)
X_test_scaled = standard_scaler.fit_transform(X_test, y_test)
````

**Predict housing price with LinearRegression, Ridge, SGDRegressor algirithms**

````python
from sklearn.linear_model import LinearRegression, Ridge, SGDRegressor
rb_boston = Ridge(alpha=0.5)
rb_boston.fit(X_train_scaled, y_train)
print("Ridge intercept boston: ", rb_boston.intercept_)
print("Ridge coef boston: ", rb_boston.coef_)
````

````python
sgd_boston = SGDRegressor(max_iter=100000, tol=1e-5, eta0=0.1, random_state=42)
sgd_boston.fit(X_train_scaled, y_train)
print("SGD intercept boston: ", sgd_boston.intercept_)
print("SGD coef boston: ", sgd_boston.coef_)
````

````python
LR_boston = LinearRegression()
LR_boston.fit(X_train, y_train)
print("LinearRegression intercept boston: ", LR_boston.intercept_)
print("LinearRegression coef boston: ", LR_boston.coef_)
````

````python
prediction_lm = LR_boston.predict(X_test_scaled)
prediction_rb = rb_boston.predict(X_test_scaled)
prediction_sgd = sgd_boston.predict(X_test_scaled)
````

````python
from sklearn import metrics
````

````python
MSE_lm = metrics.mean_squared_error(y_test, prediction_lm)
MSE_rb = metrics.mean_squared_error(y_test, prediction_rb)
MSE_sgd = metrics.mean_squared_error(y_test, prediction_sgd)
print("LR MSE:", MSE_lm)
print("Ridge MSE:", MSE_rb)
print("SGD MSE:", MSE_sgd)
````

````python
import math
````

````python
math.sqrt(MSE_sgd)
````

````python
from sklearn.model_selection import cross_val_score as cvs
````

````python
accuracy_lr = cvs(LinearRegression(), X_train_scaled, y_train, scoring='r2', cv=10)
accuracy_rb = cvs(Ridge(), X_train_scaled, y_train, scoring='r2', cv=10)
accuracy_sgd = cvs(SGDRegressor(), X_train_scaled, y_train, scoring='r2', cv=10)
print("Train data cross val mean LR:", accuracy_lr.mean())
print("Train data cross val mean Ridge:", accuracy_rb.mean())
print("Train data cross val mean SGD:", accuracy_sgd.mean())
````

# **Improving our model**

## Encoding Categorical Features

There are a few ways to handle categorical features:
* Assign them a numeric value
* Use a One-Hot Encoding (also known as creating "dummy" variables)

Here, we are going to apply a One-Hot Encoding (OHE) to our categorical variables. 

 

### One-Hot Encoding

After the transformation, each column of the resulting data set corresponds to one unique value of each original feature.

For example, suppose we have the following categorical feature with three different unique values.
```
+---------+
| Feature |
+---------+
| value_1 |
| value_2 |
| value_3 |
+---------+
```
After one-hot encoding, the data set looks like:
```
+-----------------+-----------------+-----------------+
| Feature_value_1 | Feature_value_2 | Feature_value_3 |
+-----------------+-----------------+-----------------+
|               1 |               0 |               0 |
|               0 |               1 |               0 |
|               0 |               0 |               1 |
+-----------------+-----------------+-----------------+
```

````python
X.info()
````

#### Pandas’ one hot encoder() function

* Provide the datafame and the columns you want to encod
* Return a dataframe with original columns replaced with much more columns with one-hot encodings

````python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
````

````python
ct = ColumnTransformer([('one-hot-encoder', OneHotEncoder(), ['CHAS', 'RAD'])], remainder='passthrough')
````

````python
X_oh=ct.fit_transform(X)
````

````python
X_oh.shape
````

````python
X_train_oh, X_test_oh, y_train, y_test = train_test_split(X_oh, y , test_size=0.4 , random_state=42)
````

````python
X_train_scaled_oh = standard_scaler.fit_transform(X_train_oh, y_train)
X_test_scaled_oh = standard_scaler.fit_transform(X_test_oh, y_test)
````

````python
rb_boston_oh = Ridge(alpha=0.5)
rb_boston_oh.fit(X_train_scaled_oh, y_train)
sgd_boston_oh = SGDRegressor(max_iter=100000, tol=1e-5, eta0=0.1, random_state=42)
sgd_boston_oh.fit(X_train_scaled_oh, y_train)
LR_boston_oh = LinearRegression()
LR_boston_oh.fit(X_train_oh, y_train)
````

````python
prediction_lm_oh = LR_boston_oh.predict(X_test_scaled_oh)
prediction_rb_oh = rb_boston_oh.predict(X_test_scaled_oh)
prediction_sgd_oh = sgd_boston_oh.predict(X_test_scaled_oh)
````

````python
MSE_lm_oh = metrics.mean_squared_error(y_test, prediction_lm_oh)
MSE_rb_oh = metrics.mean_squared_error(y_test, prediction_rb_oh)
MSE_sgd_oh = metrics.mean_squared_error(y_test, prediction_sgd_oh)
print("LR MSE:", MSE_lm_oh)
print("Ridge MSE:", MSE_rb_oh)
print("SGD MSE:", MSE_sgd_oh)
````

````python
accuracy_lr_oh = cvs(LinearRegression(), X_test_scaled_oh, y_test, scoring='r2', cv=10)
accuracy_rb_oh = cvs(Ridge(), X_test_scaled_oh, y_test, scoring='r2', cv=10)
accuracy_sgd_oh = cvs(SGDRegressor(), X_test_scaled_oh, y_test, scoring='r2', cv=10)
print("cross val mean LR:", accuracy_lr_oh.mean())
print("cross val mean Ridge:", accuracy_rb_oh.mean())
print("cross val mean SGD:", accuracy_sgd_oh.mean())
````
